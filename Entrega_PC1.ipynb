```python
# Eliminamos las columnas con gran cantidad de valores faltantes (ejemplo: más del 90% de valores faltantes)
porcentaje_limite = 0.90
columnas_a_eliminar = valores_faltantes[valores_faltantes > porcentaje_limite * len(df_permits)].index
df_limpio = df_permits.drop(columns=columnas_a_eliminar)

# Imputamos valores faltantes en columnas clave con métodos apropiados
# Para columnas numéricas, usamos la mediana
columnas_numericas = df_limpio.select_dtypes(include=['float64', 'int64']).columns
for col in columnas_numericas:
    df_limpio[col].fillna(df_limpio[col].median(), inplace=True)

# Para columnas categóricas, usamos la moda
columnas_categoricas = df_limpio.select_dtypes(include=['object']).columns
for col in columnas_categoricas:
    df_limpio[col].fillna(df_limpio[col].mode()[0], inplace=True)

# Revisar el resultado después de las modificaciones
df_limpio.isnull().sum().sort_values(ascending=False)
```

   - Para columnas categóricas, se han imputado los valores faltantes con el valor más frecuente.

```python
#ahora nos enfocamos en los nan debido a los datos que son False booleanos, y transformamos estos datos faltantes.
#consultado la pagina web del diccionatio del dataset:
#https://datasf.gitbook.io/datasf-dataset-explainers/dbi-building-permits-and-permit-addenda-data/building-permits-data-dictionary
#he observado que hay columnas que son booleanas, y por lo tanto, he decidido sustituir los nas por False, ya que no tiene sentido que sean True.
#las columnas son: Fire Only Permit,ADU, y Primary Address Flag
#hazme una funcion que cambie Y por True, y N por False
def change_YN(row_data):
        if row_data=='Y':
                return True
        elif row_data==True or row_data==False:
                return row_data
        else:
                return False
#aplica la funcion change_YN a las columnas Fire Only Permit,ADU, y Primary Address Flag
df_permits['Fire Only Permit']=df_permits['Fire Only Permit'].apply(change_YN)
df_permits['ADU']=df_permits['ADU'].apply(change_YN)
df_permits['Primary Address Flag']=df_permits['Primary Address Flag'].apply(change_YN)
df_permits['Site Permit']=df_permits['Site Permit'].apply(change_YN)
df_permits['Structural Notification']=df_permits['Structural Notification'].apply(change_YN)
df_permits['Voluntary Soft-Story Retrofit']=df_permits['Voluntary Soft-Story Retrofit'].apply(change_YN)
```

```python
#caso de uso 2: inferir un precio medio en caso de poner una vivienda en airbnb, y ver si es rentable o no.
#tenemos que establecer una media de precios del subset de airbnbs que esten a una distancia menor a 200 metros de la vivienda en cuestion.
#utilizaremos un csv con los listings de airbnb para no tener que esperar a la api de airbnb, y la libreria geopy para calcular las distancias.

listings = pd.read_csv('./listings.csv')

def average_price_zone(df_permits, listings, radius):
    # Iterate over each row in df_permits
    for i, permit_row in df_permits.iterrows():
        # Calculate the distance to all rows in listings
        listings['distance'] = listings.apply(
            lambda listing_row: geodesic(
                (permit_row['Location'].split('(')[1].split(')')[0].split(' ')[1], permit_row['Location'].split('(')[1].split(')')[0].split(' ')[0]),
                (listing_row['latitude'], listing_row['longitude'])
            ).km,
            axis=1
        )
        # Filter listings to only include rows where the distance is less than the radius
        nearby_listings = listings[listings['distance'] < radius]
        # Calculate the average price of the nearby listings
        avg_price =  nearby_listings['price'].replace('[\$,]', '', regex=True).astype(float).mean()
        # Assign the average price to a new column in the current row of df_permits
        df_permits.at[i, 'avg_price'] = avg_price

    return df_permits
```

Al final, es crucial que la presentación sea clara y se adapte al nivel técnico del público. Utilizar visualizaciones efectivas y ejemplos prácticos ayudará a que el equipo comprenda y aprecie el valor del modelo de datos propuesto.

#### Escenario 3
Esta semana se han incorporado al equipo dos analistas senior que están comenzando en el mundo de los datos, pero que tienen sobre todo experiencia con Excel y hojas de cálculo. ¿Cómo les presentarías el caso para que fueran más autónomos?

Para presentar el caso a los analistas senior que tienen experiencia principalmente en Excel y hojas de cálculo, es importante adaptar la comunicación y los materiales para que se ajusten a su zona de confort, a la vez que se les introduce gradualmente en conceptos más avanzados de análisis de datos. Aquí hay algunas estrategias que podrías utilizar:

1. **Introducción Familiar:**
   - **Comparaciones con Excel:** Comienza estableciendo paralelismos entre las hojas de cálculo de Excel y los conceptos de bases de datos. Por ejemplo, puedes comparar una hoja de cálculo con una tabla en una base de datos, y las fórmulas de Excel con consultas SQL.
   - **Visualizaciones:** Utiliza gráficos y tablas, herramientas con las que ya estarían familiarizados, para presentar los datos y resultados del análisis.

2. **Capacitación Paso a Paso:**
   - **Fundamentos de Bases de Datos:** Introduce conceptos básicos de bases de datos y SQL, usando ejemplos prácticos y relacionándolos con operaciones de Excel como filtros y funciones.
   - **Herramientas Amigables para Excel:** Presenta herramientas de BI como Tableau o Power BI que tienen interfaces de usuario intuitivas y permiten una transición más suave desde Excel.

3. **Workshops Prácticos:**
   - **Ejercicios Guiados:** Organiza sesiones de trabajo donde puedan practicar la importación de datos desde Excel a herramientas más avanzadas, la realización de consultas básicas de SQL y la creación de visualizaciones.
   - **Casos de Uso Reales:** Utiliza ejemplos y datasets reales del proyecto para que los ejercicios sean relevantes y atractivos.

4. **Documentación y Recursos de Aprendizaje:**
   - **Guías Paso a Paso:** Proporciona documentación detallada y guías de usuario para las nuevas herramientas y técnicas que se están introduciendo.
   - **Recursos en Línea:** Recomienda cursos en línea y tutoriales para un aprendizaje autodirigido en áreas como SQL, Python para análisis de datos, y herramientas de BI.

5. **Mentoría y Soporte Continuo:**
   - **Sesiones de Preguntas y Respuestas:** Organiza sesiones regulares de Q&A donde puedan resolver dudas y discutir problemas específicos que encuentren.
   - **Mentoría:** Asigna mentores del equipo de datos que puedan ofrecer orientación y apoyo continuo.

6. **Proyectos Graduales:**
   - **Tareas Incrementales:** Asigna pequeñas tareas o proyectos que incrementen gradualmente en complejidad, permitiéndoles aplicar y reforzar sus nuevas habilidades.

Al proporcionar un camino claro y accesible desde Excel hacia herramientas y métodos de análisis de datos más avanzados, facilitarás su transición y los ayudarás a ser más autónomos en el mundo del análisis de datos.
